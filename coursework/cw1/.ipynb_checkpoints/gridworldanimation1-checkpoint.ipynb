## Simulation

```python
def gridmove(sx, sy, a):
    # Special states
    if sx==0 and sy==1:
        return 4, 1, 10
    elif sx==0 and sy==3:
        return 2, 3, 5
    # North
    if a==0:     
        if sx==0:
            return sx, sy, -1
        else:
            return sx-1, sy, 0  
    # South     
    elif a==1:   
        if sx==4:
            return sx, sy, -1
        else:
            return sx+1, sy, 0
    # East
    elif a==2:
        if sy==4:
            return sx, sy, -1
        else:
            return sx, sy+1, 0
    # West
    elif a==3:   
        if sy==0:
            return sx, sy, -1
        else:
            return sx, sy-1, 0
```

```python
episodes = 1000
finaltime = 1000
eps = 0.1
alpha = 0.2
gamma = 0.9
q = np.zeros((5,5,4))

for j in range(episodes):
    sx = np.random.randint(0,5)
    sy = np.random.randint(0,5)
    
    for k in range(finaltime):
        # epsilon-greedy action
        if np.random.random()<eps:
            a = np.random.randint(0,4)
        else:
            a = np.argmax(q[sx,sy,:])
            
        # Next state and reward from action
        sxp, syp, reward = gridmove(sx, sy, a)
        
        # Q-learning update
        maxq = np.max(q[sxp, syp, :])
        q[sx, sy, a] = (1-alpha)*q[sx, sy, a] + alpha*(reward + gamma*maxq)
        
        # Next step override
        sx = sxp
        sy = syp
        
# Value function from q
q = np.round(q, 1)
v = np.max(q, axis=2)

plt.imshow(v)
for (i, j), z in np.ndenumerate(v):
    plt.text(j, i, z, ha='center', va='center')

plt.show()
```

```python
# Mapping of actions to vectors
dirx = [0,0,1,-1]
diry = [1,-1,0,0]

# Meshgrid and containers for the vector plot
x, y = np.meshgrid(np.linspace(0,4,5),np.linspace(0,4,5))
dirux = np.zeros((5,5,4))
dirvy = np.zeros((5,5,4))

for i in range(5):
    for j in range (5):
        
        # Max q = value function defining optimal actions
        m = np.max(q[i,j,:])
        
        # Check all entries to see if the same as max
        for a, b in enumerate(q[i,j,:]):
            if b==m:
                dirux[i,j,a]=dirx[a]
                dirvy[i,j,a]=diry[a]

for a in range(4):
    plt.quiver(x, y, dirux[:,:,a], dirvy[:,:,a], scale=15)
    
plt.xlim(-0.5,4.5)
plt.ylim(4.5,-0.5)
plt.show()
```

```python
import matplotlib.animation as animation
from pylab import *
```

```python
dpi = 100
```

```python
def ani_frame():
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_aspect('equal')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    im = ax.imshow(rand(300,300),cmap='gray',interpolation='nearest')
    im.set_clim([0,1])
    fig.set_size_inches([5,5])

    tight_layout()

    def update_img(n):
        tmp = rand(300,300)
        im.set_data(tmp)
        return im

    #legend(loc=0)
    ani = animation.FuncAnimation(fig,update_img,300,interval=30)
    writer = animation.writers['ffmpeg'](fps=30)

    ani.save('demo.mp4',writer=writer,dpi=dpi)
    return ani
```

## Q-learning movie

```python
episodes = 100
finaltime = 1000
eps = 0.1
alpha = 0.2
gamma = 0.9
q = np.zeros((5,5,4))
vpipe = np.zeros((episodes,5,5))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_aspect('equal')
    
for j in range(episodes):
    sx = np.random.randint(0,5)
    sy = np.random.randint(0,5)
    
    v = np.max(q, axis=2)
    vpipe[j,:,:] = v
    
    for k in range(finaltime):
        # epsilon-greedy action
        if np.random.random()<eps:
            a = np.random.randint(0,4)
        else:
            a = np.argmax(q[sx,sy,:])
            
        # Next state and reward from action
        sxp, syp, reward = gridmove(sx, sy, a)
        
        # Q-learning update
        maxq = np.max(q[sxp, syp, :])
        q[sx, sy, a] = (1-alpha)*q[sx, sy, a] + alpha*(reward + gamma*maxq)
        
        # Next step override
        sx = sxp
        sy = syp
        
plt.imshow(vpipe[-1,:,:])
plt.show()
```

```python
def update_v(n):
    im.set_data(vpipe[n,:,:])
    return im
```

```python
im = ax.imshow(vpipe[0,:,:])
im.set_clim([0,1])
fig.set_size_inches([5,5])

ani = animation.FuncAnimation(fig,update_v,episodes)
writer = animation.writers['ffmpeg'](fps=30)

ani.save('demo.mp4',writer=writer,dpi=dpi)
```
