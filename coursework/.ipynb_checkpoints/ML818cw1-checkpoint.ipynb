## Q1

### (a)

```python
def reward(a):
    return np.random.normal(loc=a, scale=1.0)
```

```python
nsamples = 10**3
nactions = 3
rewrange = np.linspace(-3,3,50)
meanrew = np.zeros(3)

plt.figure(figsize=(12, 4))

for i in range(nactions):
    reward_sample = [reward(i) for j in range(nsamples)]
    meanrew[i] = np.mean(reward_sample)
    
    plt.subplot(1,2,1)
    plt.hist(reward_sample, bins=50, density=True, alpha=0.5)
    plt.plot(rewrange+i, stats.norm.pdf(rewrange+i, loc=i, scale=1), '--', label='action='+str(i))
    
    plt.subplot(1,2,2)
    plt.violinplot(reward_sample, positions=[i], showmeans=True, showextrema=False)
    
plt.subplot(1,2,1)
plt.legend(loc='upper left')
plt.xlabel(r'$r$')
plt.ylabel(r'$p(r)$')
plt.subplot(1,2,2)
plt.xlabel('action')
plt.ylabel(r'$r$')
plt.show()

for i in range(nactions):
    print('Mean reward for action', i, '=', meanrew[i])
    
print('Action with max reward is', np.argmax(meanrew))
```

### (b)

For a sequence (or trajectory) of actions and rewards, we estimate the value $Q(a)$ of action $a$ as follows:

```python
def qa_estimator(actions, rewards, k):
    qa = np.zeros(k)
    numa = np.zeros(k)
    dena = np.zeros(k)
    
    for i, ai in enumerate(actions):
        numa[ai] += rewards[i]
        dena[ai] += 1
    
    for i in range(k):
        if dena[i]>0:
            qa[i] = numa[i]/dena[i]
        else:
            qa[i] = float('-inf')
            
    return qa
```

### (c)

We apply next this estimator with a greedy action selection to a given trajectory:

```python
nactions = 3
nstep = 1000

for eps in [0, 0.1, 0.3]:
    a = 0
    alist = [a]
    rewlist = [reward(a)]
    estqa = np.zeros(nactions)
    totrew = 0.0
    totrewlist = []
    
    for k in range(nstep):
        estqa = qa_estimator(alist, rewlist, nactions)
    
        if np.random.random()<eps:
            a = np.random.randint(0,3)
        else:
            a = np.argmax(estqa)
        
        rew = reward(a)
        totrew += rew
    
        alist.append(a)
        rewlist.append(rew)
        totrewlist.append(totrew/(k+1))
    
    plt.plot(range(nstep), totrewlist, label='eps='+str(eps))

plt.legend(loc='center right')
plt.xlabel('iteration')
plt.ylabel('reward')
plt.show()
```

### (c)

```python
nactions = 3
nstep = 1000

for eps in [0, 0.1, 0.3]:
    a = 0
    qa = np.zeros(nactions)
    totrew = 0.0
    totrewlist = []
    
    for n in range(nstep):    
        if np.random.random()<eps:
            a = np.random.randint(0,3)
        else:
            a = np.argmax(estqa)
        
        rew = reward(a)
        qa[a] = (1-1/(n+1))*qa[a]+1/(n+1)*rew
        totrew += rew
        totrewlist.append(totrew/(n+1))
    
    plt.plot(range(nstep), totrewlist, label='eps='+str(eps))

plt.legend(loc='upper right')
plt.xlabel('iteration')
plt.ylabel('reward')
plt.show()
    
```

### (d)

The $\varepsilon$-greedy algorithm can be seen as a learning algorithm for solving the $k$-armed bandit problem, but it is not a truly a reinforcement learning algorithm mainly because there is no system-agent interactions - there are only actions with rewards, which do not affect the state of a system. We then say that the actions are non-associative - they are not taken with respect to a context and don't influence any context. 

## Q2

### (a)

The transition matrix is

```python
Ptr = np.zeros((7,7))
for i in range(7):
    for j in range(7):
        if np.abs(i-j)==1:
            Ptr[i,j] = 0.4    
    Ptr[i,i] = 0.2
    
Ptr[0,0] = 0.6
Ptr[6,6] = 0.6              
Ptr
```

This is a doubly stochastic matrix (rows and columns sum to 1), so its stationary distribution is the uniform distribution.

The formula in vector form for calculating the value function is
$$
v = (I-\gamma P)^{-1} \rho,
$$
where $I$ is the identity matrix, $P$ is the transition probability above, and $\rho$ is the column vector containing the reward per state.

```python
rho = np.array([1, 0, 0, 0, 0, 0, 10])
```

```python
def vsol(g, P):
    return np.linalg.inv(np.identity(7)-g*P).dot(rho)
```

```python
sval = range(7)

for g in np.arange(0.,1.0,0.1):
    plt.plot(sval, vsol(g, Ptr))
    
plt.xlabel(r'$s$')
plt.ylabel(r'$v(s)$')
plt.show()
```

### (b)

We apply the iteration
$$
v_{k+1} = T(v_k)
$$
using the Bellman operator $T(v) = \rho+\gamma Pv$.

```python
def TBellman(v, g, P):
    return rho + g*P.dot(v)
```

Here's the result of 10 iterations for $\gamma=0.5$:

```python
niter = 10
g = 0.5
v = np.ones(7)
for k in range(niter):
    plt.plot(sval, v, 'b-', alpha=0.2)
    v = TBellman(v, g, Ptr)
    
plt.plot(sval, vsol(g, Ptr), 'k--')
plt.xlabel(r'$s$')
plt.ylabel(r'$v(s)$')
plt.show()
```

The iteration converges, although larger $\gamma$ values require more iterations. Repeating for different $\gamma$:

```python
niter = 20
for g in np.arange(0., 1.0, 0.1):
    v = np.ones(7)
    for k in range(niter):
        v = TBellman(v, g, Ptr) 
    plt.plot(sval, v)
    
plt.xlabel(r'$s$')
plt.ylabel(r'$v(s)$')
plt.show()
```

### (c)

For generating trajectories, we need to generate random choices for going up or down, or, for border states, for staying in those states:

```python
def next_state(s):
    r = np.random.random()
    if s==0:
        if r<0.4: s+=1 # Move up from 0
    elif s==6:
        if r<0.4: s-=1 # Move down from 6
    else:
        if r<0.4:
            s+=1       # Move up
        elif r<0.8:
            s-=1       # Move down
    return s
```

```python
# Parameters
g = 0.5
nstep = 50
slist = []

# Starting state and its reward
s = 2
slist.append(s)
reward = rho[s]

for i in range(nstep):
    s = next_state(s)
    slist.append(s)          
    reward += g**(i+1) * rho[s]

plt.plot(range(nstep+1), slist)
plt.xlabel(r'$t$')
plt.ylabel(r'$S_t$')
plt.show()
print('Final reward:', reward)
```

### (d)

The estimator of $v(s)$ is obtained by averaging the random reward of many simulated trajectories (remember the Monte Carlo course).

```python
g = 0.5
nstep = 1000
niter = 1000
rewardest = 0.0

for j in range(niter):
    s = 2 
    reward = rho[s]
    for i in range(nstep):
        s = next_state(s)               
        reward += g**(i+1) * rho[s]
    rewardest += reward
    
print('Estimated v(s):', rewardest/niter)
print('True value:', vsol(g, Ptr)[2])
```

## Q3

### (a)

The MoveLeft and MoveRight transition matrices are:

```python
PMoveLeft = np.zeros((7,7))
for i in range(7):
    for j in range(7):
        if i-j==1:
            PMoveLeft[i,j]=1.0

PMoveLeft[0,0]=1.0
PMoveLeft
```

```python
PMoveRight = np.zeros((7,7))
for i in range(7):
    for j in range(7):
        if j-i==1:
            PMoveRight[i,j]=1.0

PMoveRight[6,6]=1.0
PMoveRight
```

### (b)

We use the function defined earlier for obtaining the value function:

```python
for g in np.arange(0.0, 1.0, 0.1):
    plt.plot(sval, vsol(g, PMoveLeft), label=g)
    

plt.xlabel(r'$s$')
plt.ylabel(r'$v(s)$')
plt.title('MoveLeft value function')
plt.show()
```

Analysis:
- $v(s)$ is min at $s=5$ because it is the furthest away from the reward at $s=0$.
- $v(s)$ increases as we go to $s=0$ as it takes fewer steps to get to $s=0$.
- $v(s)$ is largest at $s=6$ where the reward is largest.
- For $\gamma=1$, $v(s)$ is the reward: 0 or 1 or 10 depending on $s$.
- For $\gamma>0$, $v(0)>\rho(0)=1$ because of the accumulated reward coming from staying at $s=0$.

```python
for g in np.arange(0.0, 1.0, 0.1):
    plt.plot(sval, vsol(g, PMoveRight))
    
plt.xlabel(r'$s$')
plt.ylabel(r'$v(s)$')
plt.title('MoveRight value function')
plt.show()
```

Analysis: For $\gamma>0$,
- $v(s)$ is min at $s=0$, the state of min reward.
- $v(s)$ increases as we go to $s=10$ as the max reward can be achieved by moving right.
- $v(s)$ is largest at $s=6$ where the reward is largest.
- For $\gamma>0$, $v(6)>\rho(6)=10$ because of the accumulated reward coming from staying at $s=10$.

```python
Pmix = 0.5*PMoveLeft+0.5*PMoveRight
for g in np.arange(0.0, 1.0, 0.1):
    plt.plot(sval, vsol(g, Pmix))
    
plt.xlabel(r'$s$')
plt.ylabel(r'$v(s)$')
plt.title('Mixed policy value function')
plt.show()
```

Analysis: Here we have the effect of being able to move left or right and to obtain, as a result, rewards by either reaching the state $s=0$ with 1 reward or the state $s=6$ with 10 reward.

### (c)

The Bellman optimality equation is
$$
v_*(s) = \max_a \mathbb{E}[R_{t+1}+v_*(s)|S_t=a,A_t=a].
$$
For our model,

- $R_{t+1}$ does not depend on $A_t$, only on $S_t$
- There are only two actions, MoveLeft and MoveRight
- The transition matrix for either action is deterministic: the state can only move up or down or stay the same. 

For Move Left, the expectation is thus
$$
\rho(s) + v_*(s')
$$
where $s'=s$ for $s=0$ and $s'=s-1$ otherwise. For MoveRight, we have instead
$$
\rho(s) + v_*(s')
$$
where $s'=s$ for $s=6$ and $s'=s+1$ otherwise. Taking the max between the two then yields $v_*(s)$.

### (d)

```python
g = 0.5
niter = 100
v = np.ones(7)
policy = np.zeros(7)
    
for k in range(niter):
    vtemp = np.copy(v)
    for s in range(7):
        if s==0:
            rewardMoveLeft = rho[s] + g*v[s]
        else:
            rewardMoveLeft = rho[s] + g*v[s-1]
            
        if s==6:
            rewardMoveRight = rho[s] + g*v[s]
        else:
            rewardMoveRight = rho[s] + g*v[s+1]
        
        if rewardMoveRight>rewardMoveLeft:
            vtemp[s] = rewardMoveRight
            policy[s] = 1
        else:
            vtemp[s] = rewardMoveLeft
            policy[s] = -1
        
    v = vtemp
    plt.plot(sval, v, 'b-', alpha=0.2)
    
plt.plot(sval, vsol(g, PMoveLeft), 'k--', label='MoveRight')
plt.plot(sval, vsol(g, PMoveRight), 'r--', label='MoveLeft')
plt.legend(loc="upper left")
plt.xlabel(r'$s$')
plt.ylabel(r'$v_*(s)$')
plt.show()

print('Optimal value function:', v)
print('Optimal policy', policy)
print('MoveLeft value function:', vsol(g, PMoveLeft))
print('MoveRight value function:', vsol(g, PMoveRight))
```

Analysis: The optimal value function comes from a trade-off between the MoveLeft value function and the MoveRight value function. MoveLeft is more optimal than MoveRight for states 0 and 1, as shown in the optimal policy, whereas MoveRight is more optimal than MoveLeft for all the other states.

The in-place version of the code follows by deleting the reference to vtemp:
